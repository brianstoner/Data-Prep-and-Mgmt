### Question 2
- *Create a new DataFrame that includes employees from the “DEPARTMENT OF EDUCATION ADMIN” agency where the variables are “First Name”, “Last Name”, “Title Description”, “Base Salary”, and “Total OT Paid”. Additionally, include a new variable “Total Compensation” which is the sum of “Base Salary” and “Total OT Paid”.*

```python
# Add a variable to the dataframe that calculates total compensation
nyc_payroll['Total Compensation'] = nyc_payroll['Base Salary'] + nyc_payroll['Total OT Paid']
# Select only the required variables from the dataframe and assign them to a new (sub) dataframe
sub_nyc_payroll = nyc_payroll[['First Name', 'Last Name', 'Title Description', 'Agency Name', 'Base Salary', 'Total OT Paid', 'Total Compensation']]
# Set the index on a new column that will be useful for further data wrangling
sub_nyc_payroll.set_index("Agency Name", inplace = True)
# Extract observations where the agency name is DEPARTMENT OF EDUCATION ADMIN
sub_nyc_payroll.loc[ "DEPARTMENT OF EDUCATION ADMIN" ]

```

### Question 3
- *Using set_index(), change the DataFrame’s index to “Last Name”, then locate the data for a specific last name, say “Doe”, and display their “Agency Name”, “Base Salary”, and “Total OT Paid”.*

```python
# Assign the dataframe to the variable 'df'
df = nyc_payroll[['Agency Name', 'Last Name', 'Base Salary', 'Total OT Paid']]
# Set the index to the 'Last Name' variable
df.set_index("Last Name", inplace = True)
# Extract the information for the employee with a last name ABURTO
df.loc[ "ABURTO" ]
```

### Question 4
- *Find the 5 employees with the highest “Regular Gross Paid” and calculate their average “OT Hours”. Also, reset the index if you have changed it previously.*


```python
# Reset the index from any previous changes
nyc_payroll.reset_index()
# Assign the dataframe to the variable 'df'
df = nyc_payroll[ ['Last Name', 'First Name', 'Regular Gross Paid', 'OT Hours'] ]
# return the 5 largest values of the 'Regular Gross Paid' variable
top = df.nlargest(5, 'Regular Gross Paid')
top
```

### Question 4 (continued)

```python
# Calculate the average of OT hours for the top 5 employees with the highest regular gross paid amounts
otavg = round(top.loc[:, 'OT Hours'].mean(), 2)
print('The average OT hours for the top 5 employees with the highest regular gross pay is {}'.format(otavg))
```

### Question 5
- *Sort the DataFrame by “Fiscal Year” and “Total Other Pay” in descending order, then set “First Name” as the index and use the loc accessor to retrieve the “Total Other Pay” for a specific first name, say “John”.*

```python
# Reset the index from previous functions
nyc_payroll.reset_index()
# Assign the dataframe to variable 'sub_nyc_payroll'
sub_nyc_payroll = nyc_payroll[['First Name', 'Last Name', 'Fiscal Year', 'Total Other Pay']]
df = sub_nyc_payroll.sort_values(by= ['Fiscal Year', 'Total Other Pay'], ascending=False)
df.set_index("First Name", inplace = True)
df.loc[ "JOHN" ]
```
