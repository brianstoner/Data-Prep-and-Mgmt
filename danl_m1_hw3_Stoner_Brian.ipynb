### Question 2
- *Create a new DataFrame that includes employees from the “DEPARTMENT OF EDUCATION ADMIN” agency where the variables are “First Name”, “Last Name”, “Title Description”, “Base Salary”, and “Total OT Paid”. ### Question 4
- *Find the 5 employees with the highest “Regular Gross Paid” and calculate their average “OT Hours”. Also, reset the index if you have changed it previously.*


```python
# Reset the index from any previous changes
nyc_payroll.reset_index()
# Assign the dataframe to the variable 'df'
df = nyc_payroll[ ['Last Name', 'First Name', 'Regular Gross Paid', 'OT Hours'] ]
# Return the 5 largest values of the 'Regular Gross Paid' variable
top = df.nlargest(5, 'Regular Gross Paid', keep = 'all')
top
```

### Question 4 (continued)

```python
# Calculate the average of OT hours for the top 5 employees with the highest regular gross paid amounts
otavg = round(top.loc[:, 'OT Hours'].mean(), 2)
print('The average OT hours for the top 5 employees with the highest regular gross pay is {}'.format(otavg))
```

### Question 5
- *Sort the DataFrame by “Fiscal Year” and “Total Other Pay” in descending order, then set “First Name” as the index and use the loc accessor to retrieve the “Total Other Pay” for a specific first name, say “John”.*

```python
# Reset the index from previous functions
nyc_payroll.reset_index()
# Assign the dataframe to variable 'sub_nyc_payroll'
sub_nyc_payroll = nyc_payroll[['First Name', 'Last Name', 'Fiscal Year', 'Total Other Pay']]
df = sub_nyc_payroll.sort_values(by= ['Fiscal Year', 'Total Other Pay'], ascending=False)
df.set_index("First Name", inplace = True)
df.loc[ "JOHN" ]
```

### Question 6
- *Select employees who have “OT Hours” greater than 0, calculate their “OT Rate” (Total OT Paid / OT Hours), and then find the employee with the highest “OT Rate”. Display their full name and OT Rate.*

```python
# Add the new calculated variable to the dataframe
nyc_payroll['OT Rate'] = round(nyc_payroll['Total OT Paid'] / nyc_payroll['OT Hours'], 2)
df = nyc_payroll[ ['First Name', 'Last Name', 'OT Rate', 'OT Hours'] ]
# Filter out any rows where the OT hours are != 0
df = df.loc[df['OT Hours'] > 0]
df.set_index(['OT Hours'])
# Sort 'OT Rate' descending to see the highest value in the list
df.sort_values(['OT Rate'], ascending=False)
# Keep only the top result of the employee with the highest OT rate
df.nlargest(1, 'OT Rate', keep = "all")
# Verified the result was correct by running
#df.max()







```
