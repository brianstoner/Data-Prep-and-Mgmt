### Question 1
- *Find the top 5 beer markets in terms of the number of households that purchased beer.*
- *Find the top 5 beer markets in terms of the amount of total beer consumption.*




```python
beer_mkts
```

```python
# Count the number of unique households by market
df = beer_mkts
count_by_market = df.groupby('market')['hh'].nunique()
count_by_market
# Sort descending and limit to top 5 markets by household values
top_markets = count_by_market.sort_values(ascending = False).head(5)
top_markets
```

```python
# Grouping by market, find the total consumtion by ounces in the beer_floz variable
consumption_by_market = df.groupby('market')['beer_floz'].sum()
consumption_by_market
# Sort values and limit to the top 5 results
top_markets = consumption_by_market.sort_values(ascending = False).head(5)
top_markets
```

### Question 2
- *For households that purchased BUD LIGHT at least once, what fraction of households did purchase only BUD LIGHT?*

- *For households that purchased BUSCH LIGHT at least once, what fraction of households did purchase only BUSCH LIGHT?*

- *For households that purchased COORS LIGHT at least once, what fraction of households did purchase only COORS LIGHT?*

- *For households that purchased MILLER LITE at least once, what fraction of households did purchase only MILLER LITE?*

- *For households that purchased NATURAL LIGHT at least once, what fraction of households did purchase only NATURAL LIGHT?*

- *Which beer brand does have the largest proportion of such loyal consumers?*

```python
# Filter the dataframe to households that bought the product at least once
bl_filtered = df[df['brand'] == 'BUD LIGHT']
all_households = df['hh'].unique()
# Group households that appear in the foltered dataframe
only_bud_light = bl_filtered.groupby('hh').filter(lambda x: len(x) == df[df['hh'] == x.name].shape[0])
# Calculate the fraction of households that bought at least once to all households
fraction_only_bud_light = len(only_bud_light['hh'].unique()) / len(all_households)
print(f"The fraction of households that purchased only BUD LIGHT is: {fraction_only_bud_light:.2f}")
# Repeat the above code with the various values in the brand variable
bl_filtered = df[df['brand'] == 'BUSCH LIGHT']
all_households = df['hh'].unique()
only_busch_light = bl_filtered.groupby('hh').filter(lambda x: len(x) == df[df['hh'] == x.name].shape[0])
fraction_only_busch_light = len(only_busch_light['hh'].unique()) / len(all_households)
print(f"The fraction of households that purchased only BUSCH LIGHT is: {fraction_only_busch_light:.2f}")

bl_filtered = df[df['brand'] == 'COORS LIGHT']
all_households = df['hh'].unique()
only_coors_light = bl_filtered.groupby('hh').filter(lambda x: len(x) == df[df['hh'] == x.name].shape[0])
fraction_only_coors_light = len(only_coors_light['hh'].unique()) / len(all_households)
print(f"The fraction of households that purchased only COORS LIGHT is: {fraction_only_coors_light:.2f}")

bl_filtered = df[df['brand'] == 'MILLER LIGHT']
all_households = df['hh'].unique()
only_miller_light = bl_filtered.groupby('hh').filter(lambda x: len(x) == df[df['hh'] == x.name].shape[0])
fraction_only_miller_light = len(only_miller_light['hh'].unique()) / len(all_households)
print(f"The fraction of households that purchased only MILLER LIGHT is: {fraction_only_miller_light:.2f}")

bl_filtered = df[df['brand'] == 'NATURAL LIGHT']
all_households = df['hh'].unique()
only_natural_light = bl_filtered.groupby('hh').filter(lambda x: len(x) == df[df['hh'] == x.name].shape[0])
fraction_only_natural_light = len(only_natural_light['hh'].unique()) / len(all_households)
print(f"The fraction of households that purchased only NATURAL LIGHT is: {fraction_only_natural_light:.2f}")
```

```python
import numpy as np
paidsearch = pd.read_csv('https://bcdanl.github.io/data/paidsearch.csv')
```

### Question 3
- *Summarize the mean value of revenue for each group of search_stays_on and for each date.*

```python
new_df = paidsearch
# Convert the date datatype from object to datetime
new_df['date'] = pd.to_datetime(new_df['date'])
# Group by search stays on and date for the mean value of revenue
mean_revenue_summary = round(new_df.groupby(['search_stays_on', 'date'])['revenue'].mean(), 2)
mean_revenue_summary
```

### Question 4
- *Calculate the log difference between mean revenues in each group of search_stays_on. (This is the log of the average revenue in group of search_stays_on == 1 minus the log of the average revenue in group of search_stays_on == 0.)*


```python
# Calculate the mean revenue for each of the search_stays_on groups
mean_revenue = new_df.groupby('search_stays_on')['revenue'].mean()
# Find the logarithm of the mean revenue for each group
log_mean_revenue = np.log(mean_revenue)
# Calculate the log difference
log_diff = log_mean_revenue.loc[1] - log_mean_revenue.loc[0]
print(f"The log difference between mean revenues for each group of search_stays_on is: {log_diff:.2f}")
```
