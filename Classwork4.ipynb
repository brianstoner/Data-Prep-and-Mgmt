```python
import pandas as pd
```

```python
movie = pd.read_csv("https://bcdanl.github.io/data/netflix-2019.csv")
```

```python
from google.colab import data_table
data_table.enable_dataframe_formatter()
```

```python
# Question 1
movie['type'] = movie['type'].astype(object)
```

```python
movie.info()
```

```python
# Question 2
movie[ movie['title'] == 'Limitless' ]
```

```python
# Question 3
movie.query("director == 'Robert Rodriguez' & type == 'Movie'")
```

```python
# Question 4
# movie["date_added"] = pd.to_datetime(movie["date_added"])
movie.query("date_added == '2019-07-31' | director == 'Robert Altman'")
```

```python
# Question 5
people = ['Orson Welles', 'Aditya Kripalani','Sam Raimi']
directors = movie["director"].isin(people)
movie[directors]
```

```python
# Question 6
between_dates = movie["date_added"].between('2019-05-01', '2019-06-01')
movie[between_dates]
```

```python
# Question 7
movie.dropna(subset = ["director"])
movie
```

```python
# Question 8
movie.drop_duplicates(subset = ["date_added"], keep = False)
movie.sort_values(['date_added'], ascending=True)
```
